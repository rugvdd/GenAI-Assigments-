```python
# import libraries
import torch
from diffusers import StableDiffusionPipeline
```

```python
# Load stable diffusion model
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
# generate artwork
prompt = "a futuristic city floating in the sky, digital art, highly detailed"

image = pipe(prompt).images[0]

image
```

```python
# save  the artwork
image.save("creative_art.png")
```
