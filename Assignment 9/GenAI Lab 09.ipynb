```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
# generating animations frames
frames = []

for i in range(50):
    fig, ax = plt.subplots()

    x = np.sin(i * 0.1)
    y = np.cos(i * 0.1)

    ax.scatter(x, y, s=500)

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_title("AI Generated Animation")

    fig.canvas.draw()

    frame = np.array(fig.canvas.renderer.buffer_rgba())
    frames.append(frame)

    plt.close()
```

```python
# creating animation video
imageio.mimsave("animation.gif", frames, fps=10)
```
