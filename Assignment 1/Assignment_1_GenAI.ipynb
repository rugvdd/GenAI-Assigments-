```
import os

api_key = os.environ.get("COHERE_API_KEY") # Replace with your actual Cohere API Key
```

```
from langchain_cohere import ChatCohere
llm = ChatCohere()
result=llm.invoke("What should I include in my diet to stay energetic and make a productive day?")
print(result)
```

```
import cohere

co = cohere.ClientV2(api_key=api_key)

res = co.chat(
    model="command-a-03-2025",
    messages=[
        {
            "role": "user",
            "content": "Tell me about the benefits of using AI!",
        }
    ],
)

print(res.message.content[0].text)
# "The Ultimate Guide to API Design: Best Practices for Building Robust and Scalable APIs"
```
